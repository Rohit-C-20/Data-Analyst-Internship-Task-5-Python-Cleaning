```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df=df.drop(columns=['Cabin'])
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
before=df.shape[0]
df.drop_duplicates(inplace=True)
after=df.shape[0]
print("Rows before",before,"|after:",after)
```

```python
df["Age"]= df['Age'].astype(int)
df['Survived']=df['Survived'].astype('category')
```

```python
df['FamilySize']=df['SibSp']+df['Parch']
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

# Task 5: Data Cleaning Notes (Titanic Dataset)

## 1) Dataset Loading
- Loaded Titanic dataset using `pd.read_csv()`.
- Checked first 5 rows using `df.head()` to understand columns and sample values.

- Used `df.info()` to check data types and missing values.

## 4) Duplicate Removal
- Checked and removed duplicate rows using `drop_duplicates()` to avoid repeated records.

## 5) Datatype Conversion
- Converted columns into correct datatypes (example: Age into integer) for accurate analysis.

## 6) Feature Engineering (New Column)
- Created a new column `FamilySize = SibSp + Parch` to represent total family members traveling together.

## 7) Final Verification
- Verified missing values again using `df.isnull().sum()` to confirm cleaning worked.

## 8) Export Cleaned Data
- Saved the cleaned dataset as `cleaned_data.csv` using `df.to_csv()` for further analysis or modeling.
